```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s_a=np.array([75,65,73,68,72,76])
s_b=np.array([90,47,43,96,93,51])
no=np.array([1,2,3,4,5,6])
```

```python
mean=np.mean(s_a)
mean
```

```python
Mad_a=np.sum(np.abs(s_a-mean))/len(s_a)
Mad_b=np.sum(np.abs(s_b-mean))/len(s_b)
```

```python
Mad_a,Mad_b
```

```python
plt.figure(figsize=(10,3))
plt.scatter(s_a,no,label="s_a")
plt.scatter(s_b,no,label="s_b")
plt.plot([70,70,70,70,70,70],no,c="blue",label="mean")
plt.legend()
plt.show()
```
